```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import UGate

qc = QuantumCircuit(1)

# U3Gate와 동일한 UGate
theta = 1.0
phi = 0.5
lam = 0.2

qc.append(UGate(theta, phi, lam), [0])
qc.draw('mpl')
```
