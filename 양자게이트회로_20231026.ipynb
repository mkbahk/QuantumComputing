```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(1)
qc.draw('mpl')
```

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

state = Statevector.from_instruction(qc)
print(state)
plot_bloch_multivector(state)
```

```python
qc = QuantumCircuit(1, 1)
qc.measure([0], [0])
qc.draw('mpl')
```

```python
from qiskit import Aer, execute
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=1024)
result = job.result()
result.get_counts()
```

**Quantum Gates(1) - Hadamard Gate**

```python
qc = QuantumCircuit(1, 1)
qc.h(0)
qc.draw('mpl')
```

```python
state = Statevector.from_instruction(qc)
print(state)
plot_bloch_multivector(state)
```

```python
qc.measure([0],[0])
qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=1024)
result = job.result()
counts = result.get_counts()
print(counts)
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(data=counts)
```

**Quantum Gates (2) - Pauli Gates<BR>**
2.1. Pauli-X gate(NOT)

```python
qc = QuantumCircuit(1)
qc.x(0) #x축을 중심(쪽)으로 180도 회전시킴
qc.draw('mpl')
```

```python
state = Statevector.from_instruction(qc)
print(state)
plot_bloch_multivector(state)
```

**2.2 Pauli-Y gate**

```python
qc = QuantumCircuit(1)
qc.y(0) #Y축을 중심(쪽)으로 180도 회전
qc.draw('mpl')
```

```python
state = Statevector.from_instruction(qc)
print(state)
plot_bloch_multivector(state)
```

X, Y-Gate를 적용한 상태가 벡터상태는 다르지만<br>
브로호스피어 상태는 매우 비슷하게 보인다. 그런데 Hadamard Gate를 적용한 후에는 많이 달라보인다.

Difference between Pauli-X and Pauli-Y gate
