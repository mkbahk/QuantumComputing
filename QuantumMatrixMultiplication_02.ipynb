```python
import numpy as np
from qiskit import QuantumCircuit, transpile, Aer, execute
```

```python
def quantum_matrix_multiplication(a, b):
    """Quantum Matrix Multiplication

    Args:
        a: 행렬 A
        b: 행렬 B

    Returns:
        행렬 AB
    """

    m = len(a)
    n = len(a[0])
    o = len(b[0])

    # 양자 회로 생성
    qc = QuantumCircuit(m + n + o, m + n)

    # 입력 행렬 A를 큐비트에 입력
    i=0
    j=0
    for i in range(m):
        for j in range(n):
            qc.ry(a[i][j], i + n)
    i=0
    j=0
    # 입력 행렬 B를 큐비트에 입력
    for i in range(n):
        for j in range(o):
            qc.ry(b[i][j], i + m + n)
    i=0
    j=0
    k=0
    # 양자 게이트 적용
    for i in range(m):
        for j in range(o):
            for k in range(n):
                qc.cnot(i + n, i + m + n + k)

    # 출력 행렬을 클래식 레지스터에 저장
    qc.measure(i + m + n, i)

    # 양자 회로 실행
    backend = qiskit.Aer.get_backend('qasm_simulator')
    result = backend.run(qc).result()

    # 출력 행렬 계산
    output = np.zeros((m, o))
    for i in range(m):
        for j in range(o):
            output[i][j] = result.data['counts'][i][2] / (2**(m + n))

    return output

# 행렬 A
a = [[1, 2], [3, 4]]

# 행렬 B
b = [[5, 6], [7, 8]]

# 행렬 AB
output = quantum_matrix_multiplication(a, b)

print(output)
```
