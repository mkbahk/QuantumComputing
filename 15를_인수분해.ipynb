```python
from qiskit import QuantumCircuit, Aer, execute

# 양자회로 초기화
n = 4
m = 8
qc = QuantumCircuit(n+m, n) #12 qubit과 4개의 Classcal bit를 가진 
```

```python
# 제어게이트와 측정게이트 추가
qc.h(range(n))
qc.barrier()
for i in range(n):
    qc.measure(i, i)
###for
```

```python
qc.draw('mpl')
```

```python
# 양자회로 실행
backend = Aer.get_backend('qasm_simulator')
results = execute(qc, backend, shots=1024).result()
```

```python
# 결과 출력
counts = results.get_counts()
print(counts)
```
